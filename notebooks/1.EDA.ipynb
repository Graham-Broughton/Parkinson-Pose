```python
with open('../data/Communication_all_export.json') as f:
    data = json.load(f)
```

```python
types = data['26-1'].keys()
```

```python
trail_type = data.keys()
types = data['26-1'].keys()
body_part = data['26-1']['position'].keys()
data['26-1']['position']['head']
```

```python
arr = [np.concatenate([data[y]['position'][x] for x in body_part], axis=1) for y in trail_type]
for i in range(len(arr)-1):
    while len(arr[i+1]) != len(arr[i]):
        if len(arr[i+1]) > len(arr[i]):
            np.append(arr[i], np.resize(np.array([[np.nan]]), (1, 30)), axis=0)
        else:
            np.append(arr[i+1], np.resize(np.array([[np.nan]]), (1, 30)), axis=0)
farr = np.stack(arr)
    
```

```python
arr = [np.concatenate([data[y]['position'][x] for x in body_part], axis=1) for y in trail_type]
maxl = max(len(x) for x in arr)
for i in range(len(arr)):
    l = len(arr[i])
    d = maxl - l
    if d > 0:
        arr[i] = np.concatenate((arr[i], np.tile(np.nan, (d, 30))), axis=0)
    elif d < 0:
        arr[i] = np.concatenate((arr[i], np.tile(np.nan, (-d, 30))), axis=0)
farr = np.stack(arr, axis=2)
```
